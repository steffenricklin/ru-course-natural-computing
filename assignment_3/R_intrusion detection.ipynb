### Exercise 1: Using the Negative Selection Algorithm for Language Categorization


```
install.packages("ggplot2")
```

```
#import the anomaly scores 
#install.packages("pROC")
library(pROC)
options(digits = 20)
#


roc_list = c()
for (i in seq(1,10)){
  english_output <- as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/englishr",as.character(i),".txt",sep="")))
  tagalog_output <-as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/tagalogr",as.character(i),".txt",sep="")))
  anomaly_score <- c(english_output,tagalog_output)
  label<- c(rep("english", length(english_output)),rep("tagalog",length(tagalog_output)))
  r <- c(rep(i,length(label)))
  if (i != 1){
   df <- rbind(df,data.frame(anomaly_score,label,r)) 
  }
  else{
    df <-data.frame(anomaly_score,label,r)
  }
  roc_list  = cbind(roc_list,roc( df$label[which(df$r == i)],  df$anomaly_score[which(df$r == i)], levels = c("english","tagalog"))
$auc)
}
head(df)
roc_list[4]
#AUC for r = 4: 0.79160971386914525

roc_list[1]
roc_list[9]
#AUC for r=1 : 0.54353471842536916
#and r=9 : 0.51209677419354838
plot.roc(roc( df$label[which(df$r == 1)],  df$anomaly_score[which(df$r == 1)], levels = c("english","tagalog")))
plot(roc( df$label[which(df$r == 9)],  df$anomaly_score[which(df$r == 9)], levels = c("english","tagalog")))
# ROC curve
# the x-axis showing  specificity (= false positive fraction = FP/(FP+TN))
# the y-axis showing sensitivity (= true positive fraction = TP/(TP+FN))
# As we can see, the ROC curve is slightly above the 0.5 line for both r values
# except they are above it a the opposite places. 




plot(seq(1,10),roc_list, xlab = "r-values", ylab = "AUC-values")
roc_list[3]
#The best AUC value is for r=3 with 0.83112356478950244


```

```
#hiligaynon
roc_list_h = c()
for (i in seq(1,10)){
  english_output <- as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/englishr",as.character(i),".txt",sep="")))
  other_output <-as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/hiligaynonr",as.character(i),".txt",sep="")))
  anomaly_score <- c(english_output,other_output)
  label<- c(rep("english", length(english_output)),rep("hili",length(other_output)))
  r <- c(rep(i,length(label)))
   df <- rbind(df,data.frame(anomaly_score,label,r)) 
  roc_list_h  = cbind(roc_list_h,roc( df$label[which(df$r == i)],  df$anomaly_score[which(df$r == i)], levels = c("english","hili"))
$auc)
}

#plautdietsch
roc_list_p = c()
for (i in seq(1,10)){
  english_output <- as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/englishr",as.character(i),".txt",sep="")))
  other_output <-as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/plautdietschr",as.character(i),".txt",sep="")))
  anomaly_score <- c(english_output,other_output)
  label<- c(rep("english", length(english_output)),rep("plautdietsch",length(other_output)))
  r <- c(rep(i,length(label)))
   df <- rbind(df,data.frame(anomaly_score,label,r)) 
  roc_list_p  = cbind(roc_list_p,roc( df$label[which(df$r == i)],  df$anomaly_score[which(df$r == i)], levels = c("english","plautdietsch"))
$auc)
}

#middle-english
roc_list_m = c()
for (i in seq(1,10)){
  english_output <- as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/englishr",as.character(i),".txt",sep="")))
  other_output <-as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/middler",as.character(i),".txt",sep="")))
  anomaly_score <- c(english_output,other_output)
  label<- c(rep("english", length(english_output)),rep("m_english",length(other_output)))
  r <- c(rep(i,length(label)))
   df <- rbind(df,data.frame(anomaly_score,label,r)) 
  roc_list_m  = cbind(roc_list_m,roc( df$label[which(df$r == i)],  df$anomaly_score[which(df$r == i)], levels = c("english","m_english"))
$auc)
}

#xhosa
roc_list_x = c()
for (i in seq(1,10)){
  english_output <- as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/englishr",as.character(i),".txt",sep="")))
  other_output <-as.numeric(readLines(paste("./ru-course-natural-computing/assignment_3/data/xhosar",as.character(i),".txt",sep="")))
  anomaly_score <- c(english_output,other_output)
  label<- c(rep("english", length(english_output)),rep("xhosa",length(other_output)))
  r <- c(rep(i,length(label)))
   df <- rbind(df,data.frame(anomaly_score,label,r)) 
  roc_list_x  = cbind(roc_list_x,roc( df$label[which(df$r == i)],  df$anomaly_score[which(df$r == i)], levels = c("english","xhosa"))
$auc)
}
```
